```python
import pandas as pd
```

```python
# Reading all input files
accounts = pd.read_csv("../data/raw/ravenstack_accounts.csv")
subscriptions = pd.read_csv("../data/raw/ravenstack_subscriptions.csv")
usage = pd.read_csv("../data/raw/ravenstack_feature_usage.csv")
tickets = pd.read_csv("../data/raw/ravenstack_support_tickets.csv")
churn = pd.read_csv("../data/raw/ravenstack_churn_events.csv")
```

```python
tables = {
    "accounts": accounts,
    "subscriptions": subscriptions,
    "feature_usage": usage,
    "support_tickets": tickets,
    "churn_events": churn,
}
```

```python
# 1) Shapes + column names
for name, df in tables.items():
    print(f"\n{name}: {df.shape[0]:,} rows x {df.shape[1]} cols")
    print(df.columns.tolist())
```

```python
# 2) Missing values (top offenders)
for name, df in tables.items():
    na = df.isna().sum()
    na = na[na > 0].sort_values(ascending=False)
    print(f"\n{name} missing values (only >0 shown):")
    display(na.head(15))
```

```python
# 3) Parse dates (we are not transforming yet, just typing correctly)
accounts["signup_date"] = pd.to_datetime(accounts["signup_date"], errors="coerce")

subscriptions["start_date"] = pd.to_datetime(subscriptions["start_date"], errors="coerce")
subscriptions["end_date"] = pd.to_datetime(subscriptions["end_date"], errors="coerce")

usage["usage_date"] = pd.to_datetime(usage["usage_date"], errors="coerce")

tickets["submitted_at"] = pd.to_datetime(tickets["submitted_at"], errors="coerce")
tickets["closed_at"] = pd.to_datetime(tickets["closed_at"], errors="coerce")

churn["churn_date"] = pd.to_datetime(churn["churn_date"], errors="coerce")

# Quick check: any date parse failures?
date_cols = {
    "accounts": ["signup_date"],
    "subscriptions": ["start_date", "end_date"],
    "feature_usage": ["usage_date"],
    "support_tickets": ["submitted_at", "closed_at"],
    "churn_events": ["churn_date"],
}

for name, cols in date_cols.items():
    df = tables[name]
    for c in cols:
        bad = df[c].isna().sum()
        print(f"{name}.{c}: {bad} unparsable / null")
```

```python
# 4) ID uniqueness sanity checks
print("accounts.account_id unique:", accounts["account_id"].is_unique)
print("subscriptions.subscription_id unique:", subscriptions["subscription_id"].is_unique)
print("feature_usage.usage_id unique:", usage["usage_id"].is_unique)
print("support_tickets.ticket_id unique:", tickets["ticket_id"].is_unique)
print("churn_events.churn_event_id unique:", churn["churn_event_id"].is_unique)
```

```python
# 5) Referential integrity checks (do foreign keys point to valid parents?)

# subscriptions -> accounts
missing_accounts_in_subs = set(subscriptions["account_id"]) - set(accounts["account_id"])
print("subscriptions.account_id missing in accounts:", len(missing_accounts_in_subs))

# churn_events -> accounts
missing_accounts_in_churn = set(churn["account_id"]) - set(accounts["account_id"])
print("churn_events.account_id missing in accounts:", len(missing_accounts_in_churn))

# support_tickets -> accounts
missing_accounts_in_tickets = set(tickets["account_id"]) - set(accounts["account_id"])
print("support_tickets.account_id missing in accounts:", len(missing_accounts_in_tickets))

# feature_usage -> subscriptions
missing_subs_in_usage = set(usage["subscription_id"]) - set(subscriptions["subscription_id"])
print("feature_usage.subscription_id missing in subscriptions:", len(missing_subs_in_usage))
```

```python
# 6) Timeline coverage (to understand the project time span)
print("accounts signup range:", accounts["signup_date"].min(), "->", accounts["signup_date"].max())
print("subscriptions start range:", subscriptions["start_date"].min(), "->", subscriptions["start_date"].max())
print("usage date range:", usage["usage_date"].min(), "->", usage["usage_date"].max())
print("tickets submitted range:", tickets["submitted_at"].min(), "->", tickets["submitted_at"].max())
print("churn date range:", churn["churn_date"].min(), "->", churn["churn_date"].max())
```
